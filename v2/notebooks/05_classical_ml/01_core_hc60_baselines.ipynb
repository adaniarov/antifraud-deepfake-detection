# 01 — HC60 classical baselines (19 runs)

Только `hc60_*`, без TF-IDF. Метрики: val, test_seen, **test_claude_binary**, one-class `test_claude_only`.

Результаты: `v2/outputs/tables/hc60_v2/baseline_all_runs.csv` + summary figures.

```python
%matplotlib inline
from __future__ import annotations
import json
import sys
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

def resolve_v2() -> Path:
    cur = Path.cwd().resolve()
    for p in [cur, *cur.parents]:
        if (p / "v2" / "pyproject.toml").is_file():
            return p / "v2"
        if p.name == "v2" and (p / "pyproject.toml").is_file():
            return p
    raise FileNotFoundError("v2")

BASE = resolve_v2()
sys.path.insert(0, str(BASE))
from src.hc60_baseline_suite import run_all

PQ = BASE / "data" / "interim" / "features" / "core_hc60_v2.parquet"
OUT = BASE / "outputs" / "tables" / "hc60_v2"
FIG = BASE / "outputs" / "figures" / "hc60_v2"
OUT.mkdir(parents=True, exist_ok=True)
FIG.mkdir(parents=True, exist_ok=True)

df = pd.read_parquet(PQ)
res = run_all(df)
res.to_csv(OUT / "baseline_all_runs.csv", index=False)
res.to_json(OUT / "baseline_all_runs.json", orient="records", indent=2)
print(res[["exp_id", "name", "val__roc_auc", "test_seen__roc_auc", "test_claude_binary__roc_auc"]])
```

```python
# Кривые ROC / Precision–Recall / F1(Recall) — все 19 экспериментов на одной фигуре (по срезам val, test_seen, test_claude_binary).
# ΔROC-AUC (test_seen − test_claude_binary) — только таблица (без графика).
import numpy as np
from IPython.display import display

from sklearn.metrics import average_precision_score, precision_recall_curve, roc_auc_score, roc_curve
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from src.hc60_baseline_suite import build_experiments, _feat_range, _split_xy, _pos_proba


def _fit_pos_proba(
    df: pd.DataFrame,
    spec,
    train_m: pd.Series,
    eval_mask: pd.Series,
):
    cols = _feat_range(spec)
    Xtr, ytr = _split_xy(df, cols, train_m)
    Xe, ye = _split_xy(df, cols, eval_mask)
    if len(np.unique(ye)) < 2:
        return None, None
    est = spec.builder()
    model = Pipeline([("scaler", StandardScaler()), ("clf", est)]) if spec.use_scaler else est
    model.fit(Xtr, ytr)
    p = _pos_proba(model, Xe)
    return ye, p


def plot_roc_all(df: pd.DataFrame, eval_mask: pd.Series, title: str, fname: str) -> None:
    train_m = df["split"] == "train"
    exps = build_experiments()
    colors = plt.cm.tab20(np.linspace(0, 1, max(len(exps), 3)))
    fig, ax = plt.subplots(figsize=(8, 8))
    for i, spec in enumerate(exps):
        ye, pr = _fit_pos_proba(df, spec, train_m, eval_mask)
        if ye is None:
            continue
        auc = float(roc_auc_score(ye, pr))
        fpr, tpr, _ = roc_curve(ye, pr)
        ax.plot(
            fpr,
            tpr,
            lw=1.15,
            alpha=0.92,
            color=colors[i % len(colors)],
            label=f"{spec.name} (AUC = {auc:.3f})",
        )
    ax.plot([0, 1], [0, 1], "k--", lw=1, alpha=0.35)
    ax.set_xlim(0.0, 1.0)
    ax.set_ylim(0.0, 1.02)
    ax.set_xlabel("False Positive Rate (Positive label: 1)")
    ax.set_ylabel("True Positive Rate (Positive label: 1)")
    ax.set_title(title)
    ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), fontsize=5.5, borderaxespad=0)
    ax.grid(True, alpha=0.3)
    ax.set_aspect("equal", adjustable="box")
    plt.tight_layout()
    plt.show()
    fig.savefig(FIG / fname, dpi=150, bbox_inches="tight")
    plt.close(fig)


def plot_pr_all(df: pd.DataFrame, eval_mask: pd.Series, title: str, fname: str) -> None:
    train_m = df["split"] == "train"
    exps = build_experiments()
    colors = plt.cm.tab20(np.linspace(0, 1, max(len(exps), 3)))
    fig, ax = plt.subplots(figsize=(8, 8))
    for i, spec in enumerate(exps):
        ye, pr = _fit_pos_proba(df, spec, train_m, eval_mask)
        if ye is None:
            continue
        prec, rec, _ = precision_recall_curve(ye, pr)
        ap = float(average_precision_score(ye, pr))
        ax.plot(
            rec,
            prec,
            lw=1.15,
            alpha=0.92,
            color=colors[i % len(colors)],
            label=f"{spec.name} (AP = {ap:.3f})",
        )
    ax.set_xlim(0.0, 1.02)
    ax.set_ylim(0.0, 1.02)
    ax.set_xlabel("Recall (Positive label: 1)")
    ax.set_ylabel("Precision (Positive label: 1)")
    ax.set_title(title)
    ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), fontsize=5.5, borderaxespad=0)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    fig.savefig(FIG / fname, dpi=150, bbox_inches="tight")
    plt.close(fig)


def plot_f1_vs_recall_all(df: pd.DataFrame, eval_mask: pd.Series, title: str, fname: str) -> None:
    train_m = df["split"] == "train"
    exps = build_experiments()
    colors = plt.cm.tab20(np.linspace(0, 1, max(len(exps), 3)))
    fig, ax = plt.subplots(figsize=(8, 8))
    for i, spec in enumerate(exps):
        ye, pr = _fit_pos_proba(df, spec, train_m, eval_mask)
        if ye is None:
            continue
        prec, rec, _ = precision_recall_curve(ye, pr)
        denom = prec + rec
        f1 = np.where(denom > 0, 2 * prec * rec / denom, 0.0)
        f1m = float(np.nanmax(f1))
        ax.plot(
            rec,
            f1,
            lw=1.15,
            alpha=0.92,
            color=colors[i % len(colors)],
            label=f"{spec.name} (max F1 = {f1m:.3f})",
        )
    ax.set_xlim(0.0, 1.02)
    ax.set_ylim(0.0, 1.02)
    ax.set_xlabel("Recall (Positive label: 1)")
    ax.set_ylabel("F1 (from precision & recall along threshold sweep)")
    ax.set_title(title)
    ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), fontsize=5.5, borderaxespad=0)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    fig.savefig(FIG / fname, dpi=150, bbox_inches="tight")
    plt.close(fig)


EVAL_SLICES: list[tuple[pd.Series, str]] = [
    (df["split"] == "val", "val"),
    (df["core_eval_slice"] == "test_seen", "test_seen"),
    (df["core_eval_slice"] == "test_claude_binary", "test_claude_binary"),
]

for mask, tag in EVAL_SLICES:
    plot_roc_all(
        df,
        mask,
        f"ROC — все эксперименты ({tag})",
        f"roc_all_{tag}.png",
    )
    plot_f1_vs_recall_all(
        df,
        mask,
        f"F1 vs recall — все эксперименты ({tag})",
        f"f1_vs_recall_all_{tag}.png",
    )
    plot_pr_all(
        df,
        mask,
        f"Precision–recall — все эксперименты ({tag})",
        f"pr_all_{tag}.png",
    )

res = res.copy()
res["delta_roc_seen_minus_cb"] = res["test_seen__roc_auc"] - res["test_claude_binary__roc_auc"]
delta_tbl = res[
    [
        "exp_id",
        "name",
        "test_seen__roc_auc",
        "test_claude_binary__roc_auc",
        "delta_roc_seen_minus_cb",
    ]
].sort_values("delta_roc_seen_minus_cb", ascending=False)
display(delta_tbl.reset_index(drop=True))
delta_tbl.to_csv(OUT / "delta_roc_seen_minus_claude_binary.csv", index=False)
```

```python
# Leaderboard style: x = test_claude_binary ROC, y = test_seen ROC
plt.figure(figsize=(7, 6))
plt.scatter(res["test_claude_binary__roc_auc"], res["test_seen__roc_auc"], c=res["exp_id"], cmap="viridis")
for _, r in res.iterrows():
    plt.annotate(str(int(r["exp_id"])), (r["test_claude_binary__roc_auc"], r["test_seen__roc_auc"]), fontsize=7)
plt.xlabel("test_claude_binary ROC-AUC")
plt.ylabel("test_seen ROC-AUC")
plt.title("Leaderboard: robustness vs in-domain")
plt.grid(True, alpha=0.3)
plt.show()
plt.savefig(FIG / "scatter_leaderboard_seen_vs_claude_binary.png", dpi=150, bbox_inches="tight")
```

```python
# Per-family robustness heatmap (best model by mean ROC across slices)
# Pick exp_id 10 xgb_richer if present else 1
pick = 10 if 10 in res["exp_id"].values else 1
# Re-fit that spec on train, score per scenario_family on test_seen and test_claude_binary
from src.hc60_baseline_suite import build_experiments, _feat_range, _split_xy, _pos_proba
from sklearn.metrics import roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

spec = [s for s in build_experiments() if s.exp_id == pick][0]
cols = _feat_range(spec)
train_m = df["split"] == "train"
Xtr, ytr = _split_xy(df, cols, train_m)
est = spec.builder()
model = Pipeline([("scaler", StandardScaler()), ("clf", est)]) if spec.use_scaler else est
model.fit(Xtr, ytr)
mat = []
for sf in sorted(df["scenario_family"].dropna().unique()):
    for sl in ["test_seen", "test_claude_binary"]:
        m = (df["scenario_family"] == sf) & (df["core_eval_slice"] == sl)
        if m.sum() < 10 or df.loc[m, "label"].nunique() < 2:
            auc = np.nan
        else:
            X, y = _split_xy(df, cols, m)
            p = _pos_proba(model, X)
            auc = roc_auc_score(y, p) if len(np.unique(y)) > 1 else np.nan
        mat.append({"scenario_family": sf, "slice": sl, "roc_auc": auc})
pm = pd.DataFrame(mat).pivot(index="scenario_family", columns="slice", values="roc_auc")
plt.figure(figsize=(8, 5))
sns.heatmap(pm, annot=True, fmt=".3f", cmap="RdYlGn", vmin=0.5, vmax=1.0)
plt.title(f"Per-family ROC-AUC (model exp_id={pick})")
plt.show()
plt.savefig(FIG / "heatmap_per_family_robustness.png", dpi=150, bbox_inches="tight")
```
